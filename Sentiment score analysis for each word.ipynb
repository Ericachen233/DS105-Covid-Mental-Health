```python
import pandas as pd
import numpy as np
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
```

```python
import nltk 
import csv

Twitter_data = []
with open('/Users/chensheng/Downloads/DS105/mid-pandemic-final1.csv', 'r') as csv_file: 
    csv_reader = csv.reader(csv_file)
    for i in csv_reader:
        Twitter_data.append(i[1])
```

```python
def lowercase(word):
    return word.lower()

    
def processSentences(sent):
    return list(map(lowercase, sent))
```

```python
import re
sentences = []
for i in Twitter_data: 
    res = re.findall(r'\w+', i)
    sentences.append(res)
sentences = list(map(processSentences, sentences))
sentences
```

```python
for line in sentences:
    for i in line:
        if i in list(stop) or len(i) < 2:
            line.remove(i)
```

```python
subset_depress = []
subset_anxious = []
subset_mental = []
subset_therapy = []
subset_wb = []
subset_stress = []
subset_sc = []
subset_sl = []
wrd_lst = ['depress', 'anxious', 'mental', 'therapy', 'wellbeing', 'stress', 'selfcare', 'self love']

```

```python
for line in sentences:
    for i in line:
        if i == 'depress' or i == 'depressed' or i == 'depressive' or i == 'depression' or i == 'depressing':
            subset_depress.append(line)
```

```python
for line in sentences:
    for i in line:
        if i == 'wellbeing' or i == 'well-being':
            subset_wb.append(line)
```

```python
for line in sentences:
    for i in line: 
        if i == 'selfcare' or i == 'self-care':
            subset_sc.append(line)
```

```python
for line in sentences:
    for i in line: 
        if i == 'selflove' or i == 'self-love':
            subset_sl.append(line)
```

```python
for line in sentences:
    for i in line:
        if i == 'mental':
            subset_mental.append(line)
```

```python
for line in sentences:
    for i in line:
        if i == 'therapy':
            subset_therapy.append(line)
      
```

```python
for line in sentences:
    for i in line:
        if i == 'stress':
            subset_stress.append(line)
```

```python
affin = open('/Users/chensheng/Downloads/DS105/AFINN_english.txt')
sentimentScores = {}
for line in affin:
    try:
        term, score = line.split("\t")
        sentimentScores[term] = float(score)
    except:
        pass
```

```python
depress_sentiment = []
for sent in subset_depress:
	score = 0
	for word in sent:
		try:
			score += sentimentScores[word]
		except:
			pass

	s = ' '.join(sent)
	depress_sentiment.append(score)

pd.DataFrame(depress_sentiment)
print(np.amin(depress_sentiment))
print(np.amax(depress_sentiment))
print(np.sum(depress_sentiment))
print(np.mean(depress_sentiment))
print(np.var(depress_sentiment))
```
